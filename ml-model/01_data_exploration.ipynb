```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('advanced_cybersecurity_data.csv')

anomalies = df['Anomaly_Flag'].value_counts()
xlabel = ['Normal', 'Anomaly']

print(anomalies)

plt.bar(xlabel, anomalies)
plt.yticks([490, 2000, 4000, 6000, 8000, 9510])
plt.title('Anomaly Flag')
plt.xlabel('Type')
plt.ylabel('Num of adresses')

plt.show()
```
